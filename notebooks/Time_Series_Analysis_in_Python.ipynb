###**Régressions linéaires simples**

####**1. Régressions linéaires simples**
+ Dans cette partie, vous découvrirez les régressions linéaires simples de séries chronologiques.

####**2. Qu’est-ce qu’une régression?**

+ ***Une régression linéaire simple trouve `la pente, bêta, et l’interception, alpha`, d’une droite qui correspond le mieux entre une variable dépendante, $y$, et une variable indépendante, $x$. Les $x$ et les $y$ peuvent être deux séries chronologiques.***

####**3. Qu’est-ce qu’une régression?**

+ ***`Une régression linéaire est également connue sous le nom de moindres carrés ordinaires, ou MCO, car elle minimise la somme des distances au carré entre les points de données et la droite de régression`***.

####**4. Paquets Python pour effectuer des régressions**
+ ***Les techniques de régression sont très courantes, et il existe donc de nombreux paquets en Python qui peuvent être utilisés. Dans les `statsmodels, il y a OLS`. Dans numpy, il y a `polyfit`, et si vous définissez degré égal à 1, il ajuste les données à une ligne, qui est une régression linéaire. `Pandas a une méthode ols`, et `scipy a une fonction de régression linéaire`.Sachez que l’ordre de $x$ et $y$ n’est pas cohérent entre les packages. Tous ces paquets sont très similaires, et dans ce cours, vous utiliserez les statsmodels OLS.***

####**5. Exemple : Régression des rendements des petites capitalisations sur les grandes capitalisations**

+ ***Maintenant, vous allez régresser les rendements des actions à petite capitalisation sur les rendements des actions à grande capitalisation. Calculez les retours à partir des prix en utilisant la méthode `« pct_change »` dans les pandas. Vous devez ajouter une colonne de uns en tant que variable dépendante de droite. La raison pour laquelle vous devez le faire est que la fonction de régression suppose que s’il n’y a pas de colonne constante, vous souhaitez exécuter la régression sans interception. En ajoutant une colonne de uns, statsmodels calculera également le coefficient de régression de cette colonne, ce qui peut être interprété comme l’interception de la ligne. La méthode statsmodels « add constant » est un moyen simple d’ajouter une constante.***

La méthode `« fit »` exécute la régression et les résultats sont enregistrés dans une instance de classe appelée `results`.***

####**9. Relation entre R-carré et corrélation**
+ ***À partir des diagrammes de dispersion, vous avez vu que la corrélation mesure à quel point les données sont regroupées le long d’une ligne.  Le $R^2$ `mesure également dans quelle mesure la droite de régression linéaire s’adapte aux données`. Donc, comme on peut s’y attendre, il existe une relation entre `la corrélation` et le $R^2$. L’amplitude de la corrélation est la racine carrée du $R^2$. Et le signe de la corrélation est le signe de la pente de la droite de régression. Si la droite de régression est inclinée positivement, la corrélation est positive et si la droite de régression est inclinée négativement, la corrélation est négative. Dans l’exemple que vous venez d’analyser, des actions à grande et à petite capitalisation, le $R^2$ était de 0 .753, la pente de la régression était positive, donc la corrélation est alors positive la racine carrée de 0.753, ou 0 point-868, ce qui peut être vérifié en calculant directement la corrélation.***

La meilleure estimation du prix de demain est simplement le prix d'aujourd'hui.***

####**4. Qu'est-ce qu'une marche aléatoire ?**

+ ***Dans une marche aléatoire avec dérive, les prix dérivent en moyenne de $\mu$ à chaque période. Et la variation du prix pour une marche aléatoire avec dérive est toujours un bruit blanc, mais avec une moyenne de $\mu$. Donc, si nous considérons maintenant les prix des actions comme une marche aléatoire avec dérive, les rendements sont toujours du bruit blanc, mais avec un rendement moyen de $\mu$ au lieu de 0.***

####**5. Test statistique de la marche aléatoire**

+ ***Pour tester si une série comme les prix des actions suit une marche aléatoire, vous pouvez régresser les prix actuels sur les prix retardés. Si le coefficient de pente, $\beta$, n'est pas significativement différent de un, alors nous ne pouvons pas rejeter l'hypothèse nulle selon laquelle la série est une marche aléatoire.
Cependant, si le coefficient de pente est significativement inférieur à 1, nous pouvons rejeter l'hypothèse nulle selon laquelle la série est une marche aléatoire.***

####**6. Test statistique de la marche aléatoire**
+ ***Une manière identique de faire ce test est de régresser la différence de prix sur le prix décalé, et au lieu de tester si le coefficient de pente est égal à 1, nous testons maintenant s'il est égal à zéro.***

####**7. Test statistique de marche aléatoire**

+ ***Il s'agit du test de `"Dickey-Fuller"`. Si vous ajoutez d'autres prix décalés sur le côté droit, `il s'agit du test de Dickey-Fuller augmenté`.***

####**8. Test ADF en Python**
+ ***statsmodels possède une fonction, adfuller, permettant d'effectuer le test de Dickey-Fuller augmenté***.

####**9. Exemple : L'indice S&P500 est-il une marche aléatoire ?**

###**Stationnarité**

####**1. Stationnarité**
+ Il existe différentes façons de définir la stationnarité,

####**2. Qu’est-ce que la stationnarité?**
+ ***Mais dans son sens le plus strict, cela signifie que `la distribution conjointe des observations ne dépend pas du temps`. Une version moins restrictive de la stationnarité, et plus facile à tester, est la `stationnarité faible`, ce qui signifie simplement que `la moyenne, la variance et les autocorrélations des observations ne dépendent pas du temps`. En d’autres termes, pour l’autocorrélation, la corrélation entre $X_t$ et $X_{t-\tau}$ n’est qu’une fonction du lag $\tau$, et non une fonction du temps.***

####**3. Pourquoi nous en soucions-nous?**

+ ***Si un processus n’est pas stationnaire, il devient difficile à modéliser. La modélisation implique l’estimation d’un ensemble de paramètres, et si un processus n’est pas stationnaire et que les paramètres sont différents à chaque moment, il y a trop de paramètres à estimer. Vous risquez d’avoir plus de paramètres que de données réelles! La stationnarité est donc nécessaire pour un modèle parcimonieux, avec un ensemble plus petit de paramètres à estimer.***

####**4. Exemples de séries non stationnaires**
+ ***Une marche aléatoire est un type courant de série non stationnaire. La variance augmente avec le temps. Par exemple, si les cours des actions sont une marche aléatoire, alors l’incertitude sur les prix de demain est beaucoup moins grande que l’incertitude dans 10 ans.***

####**1. Présentation d'un modèle AR**

+ ***Dans un modèle autorégressif, ou modèle AR, la valeur d'aujourd'hui est égale à une moyenne plus une fraction phi de la valeur d'hier, plus le bruit.***

####**2. Description mathématique du modèle AR(1)**

+ ***Puisqu'il n'y a qu'une seule valeur retardée sur le côté droit, cela s'appelle un modèle AR d'ordre 1, ou simplement un modèle AR (1). Si le paramètre AR, phi, est un, alors le processus est une marche aléatoire. Si phi est égal à zéro, le processus est un bruit blanc. Pour que le processus soit stable et stationnaire, phi doit être compris entre -1 et +1.***

####**3. Interprétation du paramètre AR(1)**

+ ***A titre d'exemple, supposons que $R_t$ est une série chronologique de rendements boursiers. Si phi est négatif, alors un rendement positif sur la dernière période, au temps $t-1$, implique que le rendement de cette période est plus susceptible d'être négatif. Cela a été appelé « retour à la moyenne » au chapitre 1. D'autre part, si phi est positif, alors un rendement positif au cours de la dernière période implique que le rendement de cette période devrait être positif. C'est ce que l'on a appelé le "momentum" au chapitre 1.***

####**4. Comparaison des séries chronologiques AR(1)**

+ ***Voici quatre séries temporelles simulées avec différents paramètres AR. Lorsque phi est égal à 0-point-9, cela ressemble à une marche aléatoire. Lorsque phi est égal à moins 0-point-9, le processus semble plus erratique - une grande valeur positive est généralement suivie d'une grande valeur négative. Les deux derniers sont similaires, mais sont moins exagérés et plus proches du bruit blanc.***

####**5. Comparaison des fonctions d'autocorrélation AR(1)**

+ ***Voici quatre fonctions d'autocorrélation pour différents paramètres $AR$. L'autocorrélation décroît exponentiellement à un taux de phi. Par conséquent, si phi est de 0 point 9, l'autocorrélation du décalage 1 est de 0 point 9, l'autocorrélation du décalage 2 est de 0 point 9 au carré, l'autocorrélation du décalage 3 est de 0 point 9 au cube, etc. Lorsque phi est négatif, la fonction d'autocorrélation décroît toujours de façon exponentielle, mais les signes de la fonction d'autocorrélation s'inversent à chaque décalage.***

####**2. Description mathématique du modèle MA(1)**
+ ***la valeur d'aujourd'hui est égale à une moyenne plus le bruit, plus une fraction thêta du bruit d'hier. Puisqu'il n'y a qu'une seule erreur retardée sur le côté droit, cela s'appelle un modèle MA d'ordre 1, ou simplement un modèle MA (1). Si le paramètre MA, thêta, est égal à zéro, alors le processus est un bruit blanc. Les modèles MA sont stationnaires pour toutes les valeurs de thêta.***

####**3. Interprétation du paramètre MA(1)**
+ ***Supposons que $R_t$ soit une série chronologique de rendements boursiers. Si thêta est négatif, alors un choc positif au cours de la dernière période, représenté par epsilon t-1, aurait rendu le rendement de la dernière période positif, mais le rendement de cette période est plus susceptible d'être négatif. Un choc il y a deux périodes n'aurait aucun effet sur le rendement d'aujourd'hui - seulement le choc maintenant et la dernière période. Notez également que l'autocorrélation lag-1 s'avère ne pas être thêta, mais thêta sur 1 plus thêta au carré.***

####**4. Comparaison des fonctions d'autocorrélation $MA(1)$**
+ ***Voici quatre fonctions d'autocorrélation pour différents paramètres $MA$. Dans chaque cas, il n'y a aucune autocorrélation pour une $MA(1)$ au-delà de lag-1. Lorsque thêta est positif, l'autocorrélation lag-1 est positive et lorsque thêta est négatif, l'autocorrélation lag-1 est négative.***

####**5. Exemple de processus $MA(1)$ : rendements boursiers intrajournaliers**
+ ***Les rendements boursiers à fréquence plus élevée sont un bel exemple de processus $MA(1)$. Voici un tracé intrajournalier du cours de l'action de Sprint Corporation pour une journée. ###**6. Fonction d'autocorrélation des rendements boursiers intrajournaliers**

####**Bitcoin et Ethereum sont-ils cointégrés ?**

+ ***La cointégration implique deux étapes : régresser une série temporelle sur l'autre pour obtenir le vecteur de cointégration, puis effectuer un test ADF sur les résidus de la régression. Dans le dernier exemple, il n'était pas nécessaire d'effectuer la première étape puisque nous supposions implicitement que le vecteur de cointégration était $\small (1,-1)$. En d'autres termes, nous avons pris la différence entre les deux séries (après avoir fait une conversion d'unités). Ici, vous ferez les deux étapes.***

+ ***Vous régresserez la valeur d'une crypto-monnaie, bitcoin (BTC), sur une autre crypto-monnaie, ethereum (ETH). Si nous appelons le coefficient de régression $\small b$, alors le vecteur de cointégration est simplement $\small (1,-b)$. Effectuez ensuite le test ADF sur BTC$\small   -  b$ETH. Les prix Bitcoin et Ethereum sont préchargés dans DataFrames BTCet ETH.***

####**Instructions**

+ Importez le module statsmodels pour la régression et la fonction adfuller
+ Ajouter une constante au DataFrame ETH en utilisant sm.add_constant()
+ Régressez BTC en ETHutilisant sm.OLS($y,x$).fit(), où $y$ est la variable dépendante et $x$ est la variable indépendante, et enregistrez les résultats dans result.
+ L'intersection est dans result.params[0] et la pente dans result.params[1]
+ Exécuter le test ADF sur BTC$\small   -  b$EPF

```python
BTC = pd.read_csv('BTC.csv', index_col=0)
ETH = pd.read_csv('ETH.csv', index_col=0)
```

```python
BTC.index = pd.to_datetime(BTC.index, format='%Y-%m-%d')
ETH.index = pd.to_datetime(ETH.index, format='%Y-%m-%d')
```

```python
# Import the statsmodels module for regression and the adfuller function
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller

# Regress BTC on ETH
ETH = sm.add_constant(ETH)
result = sm.OLS(BTC,ETH).fit()

# Compute ADF
b = result.params[1]
adf_stats = adfuller(BTC['Price'] - b*ETH['Price'])
print("The p-value for the ADF test is ", adf_stats[1])
```

***Les données suggèrent que Bitcoin et Ethereum sont cointégrés.***
