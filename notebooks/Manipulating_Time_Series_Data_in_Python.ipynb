####**5. .shift(): Déplacement des données entre le passé et le futur**

+ Votre première méthode de série chronologique est le décalage de points. 

+ Il vous permet de déplacer toutes les données d’une série ou d’un DataFrame vers le passé ou le futur. 

+ **La version « décalée » du cours de l’action a tous les prix déplacés de 1 période dans le futur**. 

+ Par conséquent, la première valeur de la série est maintenant manquante.

####**2. Corrélation et relations entre les séries**

+ La corrélation est la mesure clé des relations linéaires entre deux variables.

+ Sur les marchés financiers, les corrélations entre les rendements des actifs sont importantes pour les modèles prédictifs et la gestion des risques, par exemple.

####**3. Corrélation et relations linéaires**
+ Examinons plus en détail les corrélations et les relations linéaires entre les variables. 

+ Le coefficient de corrélation examine les relations par paires entre les variables et mesure la similitude des mouvements par paires de deux variables autour de leurs moyennes respectives. 

+ Ce co-mouvement par paires est appelé covariance. 

+ Le coefficient de corrélation divise cette mesure par le produit des écarts-types pour chaque variable. 

+ En conséquence, le coefficient varie entre -1 et +1. 

+ Plus le coefficient de corrélation est proche de plus ou 1 ou moins 1, plus un tracé des paires des deux séries ressemble à une ligne droite. 

+ Le signe du coefficient implique une relation positive ou négative. 

+ Une relation positive signifie que lorsqu’une variable est au-dessus de sa moyenne, l’autre est probablement également au-dessus de sa moyenne, et vice versa pour une relation négative. 

+ Il existe cependant de nombreux types de relations non linéaires que le coefficient de corrélation ne saisit pas.
